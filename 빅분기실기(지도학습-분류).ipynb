```python
import numpy as np
import pandas as pd
import sklearn

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

```python
df
```

```python
df.shape
df.info()
df.describe()
```

```python
df["species"].replace({'setosa' : 0, 'versicolor' : 1, 'virginica' : 2}, inplace = True)
df
```

```python
X = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
y = df["species"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
dt = DecisionTreeClassifier(random_state=11)
dt.fit(X_train, y_train)
```

```python
pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
```

```python
from sklearn.metrics import classification_report
rpt = classification_report(y_test, pred)
print(rpt)
```

```python
import numpy as np
import pandas as pd
import sklearn

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
d_mean = df["Age"].mean()
df["Age"].fillna(d_mean, inplace=True)

d_mode = df["Embarked"].mode()[0]
df["Embarked"].fillna(d_mode, inplace=True)

df["FamilySize"] = df["SibSp"] + df["Parch"]
```

```python
onehot_sex = pd.get_dummies(df["Sex"])
df = pd.concat([df, onehot_sex], axis = 1)

onehot_embarked = pd.get_dummies(df["Embarked"])
df = pd.concat([df, onehot_embarked], axis = 1)
```

```python
df
```

```python
X = df[["Pclass", "Age", "Fare", "FamilySize", "female", "male", "C", "Q", "S"]]
y = df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
sv = SVC(kernel = 'rbf')
sv.fit(X_train, y_train)
```

```python
pred = sv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pred)
print(mat)
```

```python
from sklearn.metrics import classification_report
rpt = classification_report(y_test, pred)
print(rpt)
```

```python
import numpy as np
import pandas as pd
import sklearn

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
d_mean = df["Age"].mean()
df["Age"].fillna(d_mean, inplace=True)

d_mode = df["Embarked"].mode()[0]
df["Embarked"].fillna(d_mode, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
df["Sex"] = LabelEncoder().fit_transform(df["Sex"])

df["Embarked"] = LabelEncoder().fit_transform(df["Embarked"])
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"]
```

```python
X = df[["Pclass", "Sex", "Age", "Fare", "Embarked", "FamilySize"]]
y = df["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=11)
```

```python
rf = RandomForestClassifier(n_estimators = 50, max_depth = 3, random_state = 20)
rf.fit(X_train, y_train)
```

```python
pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```
